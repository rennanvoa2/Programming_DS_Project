```python
# Load the Pandas libraries with alias 'pd' 

import pandas as pd
```

```python
# Read data from file 'International Arrivals.csv' that contains for each country or zone the number of international arrivals per year

arrivals=pd.read_csv('International Arrivals.csv', header=4)

#Select only the columns needed. Exclude information from 1960 to 1994 and 2018 that is empty

arrivals=arrivals[['Country Name', 'Country Code', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017']]
```

```python
#Check the arrivals data frame

arrivals
```

```python
# Read data from file 'Metadata_Country.csv' containing further information about countries

metadata_country=pd.read_csv('Metadata_Country.csv')
```

```python
#Check the metadata_country data frame

metadata_country
```

```python
#Merge both data frames on 'Country Code' to have consolidate information an create a new data frame called df

df=arrivals.merge(metadata_country, on='Country Code', how='left')

arrivals
```

```python
#Find the top 10 countries or regions with the most number of visitor in 2017

df.nlargest(10, ['2017']) 
```

```python
#Find the top 10 countries exclusively with the most number of visitor in 2017. 
#We need to exclude Regions and Income Groups to look only for countries data

df[(df['Region'].notnull()) & (df['IncomeGroup'].notnull())].nlargest(10,['2017'])
```

```python
#Find the bottom 10 only countries exclusively with the most number of visitor in 2017. 

df[(df['Region'].notnull()) & (df['IncomeGroup'].notnull())].nsmallest(10,['2017'])
```
